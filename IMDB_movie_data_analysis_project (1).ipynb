```python
# Optional Modeling (Predicting Movie Ratings)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Prepare features and target variable

x=df[['Runtime (Minutes)','Votes','Revenue (Millions)']]
y=df['Rating']

# Split into train and test sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)

# Create and train the model
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
# Make predictions
y_predict=model.predict(x_test)
y_predict
```

```python
# Evaluate the model
mse=mean_squared_error(y_test,y_predict)
print(f'Mean Squared Error: {mse}')
```

```python
model.score(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
sns.regplot(x=y_test, y=y_predict, color='blue',line_kws={'color': 'red'})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a scatter plot with a regression line
plt.figure(figsize=(10, 6))
sns.regplot(x=y_test, y=y_predict, scatter_kws={'color': 'blue', 'alpha': 0.6}, line_kws={'color': 'red'})
plt.title('Actual vs Predicted Ratings with Regression Line')
plt.xlabel('Actual Ratings')
plt.ylabel('Predicted Ratings')
plt.show()
```
